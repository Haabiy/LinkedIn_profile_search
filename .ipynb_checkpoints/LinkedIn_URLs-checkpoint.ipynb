```python
from bs4 import BeautifulSoup
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import csv, codecs
from time import sleep
```

### <font color='Orange'> 1.1 Log in</font> 

```python
# Chrome as my web browser, signing into my Sales navigator account
driver = webdriver.Chrome(ChromeDriverManager().install())
url = 'https://www.linkedin.com/uas/login?session_redirect=/sales&fromSignIn=true&trk=navigator'
driver.get(url)
```

```python
# Locate the two boxes to enter your credentials(email and password) into and the remaining one to submit(sign in button)
email = driver.find_element(By.ID, ("username"))
password = driver.find_element(By.ID, ("password"))
login = driver.find_element(By.XPATH, '//*[@id="app__container"]/main/div[3]/form/div[3]/button')
```

```python
def login():
    # Include a .txt named Authentication
    # Index = 0 for your email and index = 1 for your password
    # In fact, you could input your email and password from the user, however, 
    # entering credentials each time when testing the script is tedious, it's doable anyway
    opencredentials = open('./Authentication.txt')
    credentials = opencredentials.readlines()
    myemail = credentials[0]
    mypassword = credentials[1]

    # First, make sure you clear the two input boxes
    email.clear()
    password.clear()

    # send_keys :  enters the credentials you specified on the text file
    # Keys.RETURN == PRESSING ENTER on the keyboard
    email.send_keys(myemail)
    password.send_keys(mypassword)
    password.send_keys(Keys.RETURN)
    #login.submit()
```

### <font color='Orange'> 1.2 Save Sales Navigator LinkedIn URLs</font> 

```python
# throws an error if the loc is just int: in order to concatenate change the int type object to a string type 
# https://stackoverflow.com/questions/51252580/how-to-resolve-typeerror-can-only-concatenate-str-not-int-to-str
# Two storages, one to store urls and the other to store names
store_urls = []
store_names = []
class get_urls():
    # _search_bar function takes 4 parameters
    # file : the name of the file
    # loc : Looking at the xpath, loc is for location, in the sales navigator account; the first name is located on loc = 6 
    # where as last name is located on loc = 7
    # title for the industry type to search for
    # i : to go through the list of industries and first names
    def _search_bar(self, file, loc, title, i):
        self.file = file
        self.loc = loc
        self.i = i
        
        # open and read the file containing first names
        lst_first_names = open('./' + self.file + '.txt', encoding='utf-8')
        read_first_names = lst_first_names.readlines()
        sleep(1)
        try:
            # for the search iterations apart from iteration = 0, we need to remove the first name before we
            # search for a new one, we need to make sure that the box is empty first
            if(self.i!=0):
                first_name = driver.find_element(By.XPATH,'//*[@id="content-main"]/div/div[1]/form/div/div[4]/fieldset[1]/div/fieldset[6]/ul/li[1]')
                first_name.click()
            
            # self.i = 0 starts from here
            # first of all, we need to clear the search bar
            # then send the key term, in our case, the type of industry -> will be passed through 'title'
            search_bar = driver.find_element(By.XPATH, '//*[@id="global-typeahead-search-input"]')
            search_bar.clear()
            search_bar.send_keys(title)
            driver.maximize_window()
            sleep(2)
            
            # for iteration i = 0, since this is the very first, we need to click on 'see all filters' option
            # before filtering them out based on their first names, next, click on the first name
            if(i==0):
                more_filters = driver.find_element(By.XPATH, '//*[@id="content-main"]/div/div[1]/form/div/div[3]/div/button')
                more_filters.click()
                first_name = driver.find_element(By.XPATH, '//*[@id="content-main"]/div/div[1]/form/div/div[4]/fieldset[1]/div/fieldset[' + str(self.loc) + ']/div/button/li-icon')
                first_name.click()
            sleep(2)
            
            # filtering out based on their first name
            fname = read_first_names[self.i]
            input_first_name = driver.find_element(By.XPATH, '//*[@id="content-main"]/div/div[1]/form/div/div[4]/fieldset[1]/div/fieldset[6]/div[3]/div/input')
            input_first_name.click()
            input_first_name.send_keys(fname)
            sleep(3)
            
            # from the displayed search results, we can retrieve their account urls
            soup = BeautifulSoup(driver.page_source, 'html.parser')
            for div in soup.find_all('div', {'class' : 'flex flex-wrap align-items-center'}):
                for links_ in range(10):
                    try:
                            name = div.find_all('a')[links_].text.strip()
                            for link_loc in div.find_all('a', href=True):
                                link = link_loc['href']
                                store_urls.append('https://www.linkedin.com'+link)
                                store_names.append(name)
                                print(name,'https://www.linkedin.com'+link)
                    except Exception:
                        pass
        except Exception:
            pass
```

### <font color='Orange'> 1.3 Export Sales Navigator URLs to .CSV format</font> 

```python
# Wait a moment before running this code for LinkedIn might send a security verification
# verify and run the following script

# list of industries to search for
sleep(3)
search_bar = driver.find_element(By.XPATH, '//*[@id="global-typeahead-search-input"]')
industry_list = open('./Industry.txt', encoding='utf-8')
title = industry_list.readlines()
```

```python
urls_ = get_urls()
for i in range(len(title)):
    industry_type = title[i]
    urls_._search_bar('First_names', 6, industry_type , i)
    sleep(2)
```

```python
# Store Sales nav LinkedIn urls in a .csv format
with codecs.open(r'Linkedin_sales_nav.csv', 'w', encoding='utf-8') as file:
    for i in range(len(store_urls)):
        if store_urls[i] not in store_urls[i + 1:]:
            #print(store_names[i])
            #print(store_urls[i])
            write_ = csv.writer(file)
            write_.writerow([store_names[i], store_urls[i]])
```

### <font color='Orange'> 1.4 Store and export LinkedIn URLs to .CSV format</font> 

```python
# our main goal is to get their LinkedIn urls, so, we can go profile by profile to retrieve what we need
linkedin_links = []
def get_linkedin_urls(urls):
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    driver.get(urls)
    sleep(3)
    try:
        # divclass is the location where the LinkedIn urls will be stored
        divclass = soup.find('div', {'class' : '_container_x5gf48 _container_iq15dg _raised_1aegh9'})
        for a in divclass.find_all('a', href = True):
            print(a['href'])
            linkedin_links.append(a['href'])
    except Exception:
        pass
```

```python
for urls_ in store_urls:
    get_linkedin_urls(urls_)
    sleep(3)
```
